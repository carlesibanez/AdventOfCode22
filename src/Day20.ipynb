# Day 20

```python
with open('../inputs/adventofcode.com_2022_day_20_input.txt', 'r') as f:
    data = f.read().splitlines()

print(f'The file contains {len(data)} numbers.')
```

## Puzzle 1

```python
original = [(int(n), i) for i, n in enumerate(data)]
new_list = original.copy()

def mixing(original, new_list):
    for n in original:
        idx = new_list.index(n)

        l1 = new_list[:idx]
        l2 = new_list[idx+1:]
        new_list = l1 + l2
        if (idx+n[0]) % len(new_list) == 0:
            new_list = new_list + [n]
        else:
            new_list = new_list[:(idx+n[0]) % len(new_list)] + [n] + new_list[(idx+n[0]) % len(new_list):]

    return new_list

new_list = mixing(original, new_list)
print(f'After mixing{new_list}')
```

```python
idx_0 = [n[0] for n in new_list].index(0)
print(f'Index of 0: {idx_0}')
list1000 = new_list[(idx_0+1000) % len(new_list)][0]
print(f'1000th element: {list1000}')
list2000 = new_list[(idx_0+2000) % len(new_list)][0]
print(f'2000th element: {list2000}')
list3000 = new_list[(idx_0+3000) % len(new_list)][0]
print(f'3000th element: {list3000}')

result = list1000 + list2000 + list3000
print(f'The sum of the 1000th, 2000th and 3000th elements is: {result}.')
```

## Puzzle 2

```python
decryption_key = 811589153
original = [(int(n)*decryption_key, i) for i, n in enumerate(data)]
new_list = original.copy()


for r in range(10):
    new_list = mixing(original, new_list)

    print(f'After round {r+1} moving: {new_list}')
```

```python
idx_0 = [n[0] for n in new_list].index(0)
print(f'Index of 0: {idx_0}')
list1000 = new_list[(idx_0+1000) % len(new_list)][0]
print(f'1000th element: {list1000}')
list2000 = new_list[(idx_0+2000) % len(new_list)][0]
print(f'2000th element: {list2000}')
list3000 = new_list[(idx_0+3000) % len(new_list)][0]
print(f'3000th element: {list3000}')

result = list1000 + list2000 + list3000
print(f'The sum of the 1000th, 2000th and 3000th elements is: {result}.')
```
